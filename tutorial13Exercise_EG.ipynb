```python
import numpy as np
import seaborn as sns
```

```python
means = [5,7,9,11,13]
sd = [2,4,6,8,10]
```

```python
distsize = 5

dist = np.zeros((distsize,100))

for i in np.arange(0,len(means)):
    dist[i,:] = np.random.randn(1,100) * sd[i] + means[i]
    
dist.shape
```

```python
for i in np.arange(0,len(means)):
    sns.kdeplot(dist[i,:])
```

```python
for i in range(len(means)):
               sns.displot(dist[:,i], kind ='kde')
```

The single axis plot and the multiple axis plot look so different because the first one is plotting all of the distributions at once and the second one is plotting them individually.

The first plot is more informative because you can see the distributions on top of each other and compare their differences directly. It is a lot harder to compare the distributions when they are plotted seperately because the differences in their curves are less apparent.

```python
n_replications = 1
mc_means = np.zeros((2000, 1))

while n_replications < 2000:
    mean = np.random.randn() * 10 + 50
    mean = np.mean(mean)
    mc_means[n_replications] = mean
    n_replications += 1
mc_means
```

```python
sns.displot(mc_means, kind='hist')
```

```python
sns.displot(mc_means, kind='kde')
```

```python
newmean = np.mean(mc_means)
print(newmean)
```

```python
originalmean = np.mean(means)
print(originalmean)
print(newmean)
```

```python
sd = np.std(mc_means, ddof=1)
print(sd)
```

```python
originalsem = np.std(means, ddof=1) / np.sqrt(np.size(means))
print(originalsem)

newsem = np.std(mc_means, ddof=1) / np.sqrt(np.size(mc_means))
print(newsem)
```
